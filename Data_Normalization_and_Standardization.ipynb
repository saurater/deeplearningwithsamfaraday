# 1. Data Normalization and Standardization

Both Normalization and Standardization are preprocessing steps we take to:

1. Reduce the size of the data. As we process data, lots of transformations are applied producing really large numbers by a series of multiplications and other Math Operations. 

If you do not reduce the size, your computer may not have enough memory to process the data and/or could take much longer.

2. Some algorithms are really sensitive to non-normalized data, so, please make sure to always normalize data.

# 2. Should I choose Normalization or Standardization?

Normalization is useful when there are no outliers.

Standardization should be used in cases where the data follows a Gaussian  (Normal) distribution.

Standardization does not get affected by outliers because there is no predefined range of transformed features.

# 3. Hands-on

## Import Libs

```
from numpy import asarray
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

## Normalization or MinMaxScaler

### Define Data
Let us suppose there 4 different ages we need to Normalize by creating an array named age with the value 10, 20, 30, 50

**Min Value** 10

**Max Value** 50

```

age = asarray([[10],
				[20],
				[30],
				[50],
				])
```

image.png

image.png

```
print(age)
```

```
# check the variable age type
type(age)
```

### Instantiate the Scaler (Min Max Scaler)

```
scaler = MinMaxScaler()
# transform 
```

### Transform and Show the Normalizaed Data

Please note that now the ages range from 0 to 1

```
scaled = scaler.fit_transform(age)

print(scaled)
```

## Standardization

### Define Data
Let us suppose there 5 different prices: 100, 8 , 50 , 88,4.

**Outliers** 4, 8

```
item_prices = asarray([[100],
				[8],
				[50],
				[88],
				[4]])
```

image.png

### Instantiate the StandardScaler

image.png

```
scaler = StandardScaler()
```

### Transform and Show the Standardized Data

Slightly different numbers are expected when using Excel and Python

```
# transform data
scaled = scaler.fit_transform(item_prices)
```

```
print(scaled)
```
